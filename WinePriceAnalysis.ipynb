```python
#Dropping 'Region 1 and Region 2' from the dataset.
ww.drop(columns=['region_1', 'region_2'], inplace= True)
ww.head()
```

```python
ww['province'].value_counts()
```

```python
#Selecting Province as Douro for further analysis as it has been noticed that Douro has maximum number of Sales.
dv = ww[ww['province']== 'Douro']
dv.head()
```

```python
#variety of grapes used in Douro Verde wines
dv['variety'].value_counts()
```

```python
# Pie chart for mostly used grape in Douro
labels = ['Portuguese Red', 'Portuguese White', 'Touriga Nacional']
sizes = [748, 182, 46]
explode = (0.1, 0.1, 0.1)  
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90, textprops={'fontsize': 12})
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.title('Best variety of Grapes used in Douro')
plt.show()
```

```python
#Checking the price for Douro province Wines
dv['price'].value_counts()
```

```python
#Naming Price Less, Average and Expensive
dv['price_type'] = dv['price'].apply(lambda value: 'Less'
if value <= 40 else 'Average'
if value <= 60 else 'Expensive')

dv['price_type'] = pd.Categorical(dv['price_type'],
categories=['Less', 'Average', 'Expensive'])
dv.head()
```

```python
dv['price_type'].value_counts()
```

```python
#Count of Pricey Wines
fig, axs = plt.subplots(ncols=1,figsize=(8,6))
palette = ["olive", "khaki", "gold"]
sns.barplot(x='price_type', y='price', data=dv, ax=axs, palette=palette, edgecolor='Black')
plt.title('Count of Pricey Wines')

plt.tight_layout()
plt.show()
plt.gcf().clear()
```

```python
#Counting Highest Wine prices in different Countries
cnt = wsd.groupby(['country'])['price'].max().sort_values(ascending=False).to_frame()[:10]
sns.barplot(x = cnt['price'], y = cnt.index, palette= 'copper')
plt.title('Highest Wine prices in different countries');
```
